```python
%sql
create or replace table
  workspace.pj_sales.tb_dim_date_gold as
  select
     cast(date_format(date_value, 'yyyyMMdd') as bigint)   as _sk_date
    ,date_value                                            as date
    ,year(date_value)                                      as year
    ,month(date_value)                                     as month
    ,day(date_value)                                       as day_of_month
    ,weekofyear(date_value)                                as week_of_year
    ,dayofweek(date_value)                                 as day_of_week
    ,concat(year(date_value), '-H',
      case
        when month(date_value) <= 6 then 1
        else 2
      end
     )                                                     as year_semester
    ,concat(year(date_value), '-Q', quarter(date_value))   as quarter_of_year
    ,date_format(date_value, 'MMM')                        as month_name
  from (
    select explode(
      sequence(
         to_date('2015-01-01')
        ,to_date('2030-12-31')
        ,interval 1 day
      ) -- Cria uma sequência entre a data inicial e a data final com contagem de um dia
    ) /* Transforma a qualquer função collection em lista */                as date_value
  )
  order by date_value asc
-- Funções collection
  
```
